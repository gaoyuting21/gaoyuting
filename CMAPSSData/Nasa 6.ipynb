```python
#slide_windows
# ws = 5

#reform the data

def Reform_data(data,ws):
    X_N = pd.DataFrame()
    for i in range(1,4):
        df = data.iloc[:,i]
        df_max = []
        df_min = []
        for j in range(data.shape[0]-6):
            if data.iloc[j,0]==data.iloc[j+ws,0]:
                df_max.append(max(df[j:j+ws]))
                df_min.append(np.min(df[j:j+ws]))
        X_N[names[i]+'max'] = df_max
        X_N[names[i]+'min'] = df_min
    for i in range(4,data.shape[1]-1):
        df = data.iloc[:,i]
        df_mean = []
        df_max = []
        df_min = []
        df_var = []
        for t in range(data.shape[0]-6):
            if data.iloc[t,0]==data.iloc[t+ws,0]:
                df_mean.append(np.mean(df[t:t+ws]))
                df_max.append(max(df[t:t+ws]))
                df_min.append(np.min(df[t:t+ws]))
                df_var.append(np.var(df[t:t+ws]))
        X_N[names[i]+'mean'] = df_mean
        X_N[names[i]+'max'] = df_max
        X_N[names[i]+'min'] = df_min
        X_N[names[i]+'var'] = df_var
    Y_n = []
    for m in range(data.shape[0]-6):
        if data.iloc[m,0]==data.iloc[m+ws,0]:
            Y_n.append(data.iloc[m+ws,20])
        
    return X_N,Y_n
X_new, Y_new = Reform_data(X,5)

```
