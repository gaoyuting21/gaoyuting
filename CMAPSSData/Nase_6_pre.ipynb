```python
cols = ['id', 'cycle' ]

# Three operational setting columns
setting_cols = ['setting' + str(i) for i in range(1,4)]
cols.extend(setting_cols)

# Twenty one sensor columns
sensor_cols = ['s' + str(i) for i in range(1,22)]
cols.extend(sensor_cols)

# The data already comes sorted, but to be on the safe we will 
# maintain an order based on engine and cycle and ascending cycle.
sort_cols = ['id','cycle']
```

```python
data_1.columns = cols
data_1.head()
```

```python
data_test.columns = cols
data_test.head()
```

```python
def rul_cal(data):
 
    rul = []
    for temp in data.groupby("id"):
        df = temp[1]
        max_cycle = df["cycle"].max()
#     print(max_cycle,"  ", df.shape[0])
        temp_list = list(range(max_cycle))
        temp_list = temp_list[::-1]
        rul = rul + temp_list
    data["rul"] = rul
    data = data.drop(columns = ['cycle'])
    data = data.drop(columns = ['id'])
    return data
```

```python
#X = data_1.drop(columns = ['cycle'])
#Y = data_1['rul']
df_train = rul_cal(data_1)
df_test = rul_cal(data_test)
```
