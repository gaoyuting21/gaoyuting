Caculate Rul 

```python
def rul_cal(data):
 
    rul = []
    for temp in data.groupby("id"):
        df = temp[1]
        max_cycle = df["cycle"].max()
        temp_list = list(range(max_cycle))
        temp_list = temp_list[::-1]
        rul = rul + temp_list
    data["rul"] = rul
    data = data.drop(columns = ['cycle'])
    return data
```

```python
df_train = rul_cal(df)
df_test = rul_cal(df_test)
```

```python
y_train = df_train['rul']
x_train = df_train.drop(columns = ['rul'])

y_test = df_test['rul']
x_test = df_test.drop(columns = ['rul'])
```

def slide-window

```python
def slideing_window(data,y_data,wl,sl,col,temp):
    '''return a 3D arrayfor x  and a 1D arry for y
    parameters:
        data: dataframe for x 
        y_data : rul
        wl: an int representing the window_lenth
        sl: an int representing the slid_lenth
        col: an int representing the number of columns
        '''
    t = 0
    q = 0
    #caculat the dimension
    for f in range(data.shape[0]-wl):
        if data.iloc[f,0] == data.iloc[f+wl,0] and q <temp:  #chose the same engine
            t = t+1
            q = q + 1
    print(t)
    
    data_array = np.zeros((t,wl,col))
#  data_array = [[[[]]]]
    y = []
    q = 0
    for i in range(data.shape[0]-wl):
        if data.iloc[i,0] == data.iloc[i+wl,0] and q <temp:  #chose the same engine
            #data_array.append(data.iloc[i+1:i+wl+1,1:col+1])
            data_array[q] = data.iloc[i+1:i+wl+1,1:col+1]
            q = q + 1
            y.append(y_data.iloc[i+wl])
    print(q)
    return data_array, y
```

```python
#reshape with def sliding_window
x_train_w,y_train_w = slideing_window(x_train,y_train,24,1,24,18000)
x_test_w,y_test_w = slideing_window(x_test,y_test,24,1,24,10000)
```

Build Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)
```

```python
model = keras.Sequential()
model.add(layers.ConvLSTM2D(filters=32, kernel_size=(3,24),return_sequences=True))
model.add(layers.BatchNormalization())
model.add(layers.ConvLSTM2D(filters=64, kernel_size=(22,1)))
#model.add(layers.Conv2D(filters = 64, kernel_size = (22,1)))
#model.add(layers.ConvLSTM2D(filters=40, kernel_size=(3,1),return_sequences=True))
model.add(layers.Flatten())
model.add(layers.Dense(100))
model.add(layers.Dense(5))
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(x_train_r, y_train_a, batch_size=10,epochs=5)
```

```python
yhat = model.predict(x_test_r)
```
